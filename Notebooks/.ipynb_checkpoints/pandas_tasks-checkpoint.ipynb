```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [19, 22, 20, 23, 18],
    'Score': [85, 78, 92, 65, 88]
}
df = pd.DataFrame(data)
print("Full DataFrame:\n", df)
```

```python
print("\nFirst 3 rows:\n", df.head(3))
print("\nSummary Statistics:\n", df.describe())
average_score = df['Score'].mean()
print(f"\nAverage Score: {average_score}")
```

```python
students_older_20 = df[df['Age'] > 20]
print("\nStudents older than 20:\n", students_older_20)
high_scorers = df[df['Score'] > 80]
print("\nStudents with score > 80:\n", high_scorers)
```
