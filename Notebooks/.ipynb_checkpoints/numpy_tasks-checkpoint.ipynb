```python
import numpy as np
arr_1d = np.arange(1, 11)
print("1D Array:", arr_1d)
arr_2d = np.random.randint(1, 100, (3, 3))
print("\n2D Array (3x3):\n", arr_2d)
```

```python
print(f"\nSum: {arr_1d.sum()}")
print(f"Mean: {arr_1d.mean()}")
print(f"Max: {arr_1d.max()}")
arr_mult = arr_1d * 2
print("\nMultiplied by 2:", arr_mult)
arr_replaced = arr_1d.copy()
arr_replaced[arr_replaced > 5] = 0
print("Values > 5 replaced with 0:", arr_replaced)
```
