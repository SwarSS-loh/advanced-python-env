```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [85, 78, 92, 65, 88]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 5)) 
plt.plot(df['Score'], marker='o', linestyle='-', color='b') 
plt.xlabel('Student Index')
plt.ylabel('Score')
plt.title('Student Scores Trend')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Score'], color='green')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Scores by Student')
plt.show()
```
